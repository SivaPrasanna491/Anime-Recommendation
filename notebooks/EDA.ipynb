```python
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
import re
```

```python
corpus = []
for i in range(df.shape[0]):
    data = str(df['englishTitle'][i])
    data = data.lower()
    data = data.split(" ")
    data = [re.sub('^a-zA', '', word) for word in data]
    data = [word for word in data if word not in stopwords.words('english')]
    data = [lemmatizer.lemmatize(word) for word in data]
    corpus.append(' '.join(data))
```

```python
corpus
```

```python
## Performing tokenization
word_tokens = []

for word in corpus:
    word_tokens.append(nltk.word_tokenize(word))
    
```

```python
word_tokens
```

```python
## Applying one_hot
import tensorflow as tf
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
vocab_size = 500
one_hot_repr = [one_hot(word, vocab_size) for word in corpus]
```

```python
one_hot_repr
```

```python
## Apply padding 
from tensorflow.keras.utils import pad_sequences
```

```python
max_length = 0
for token in word_tokens:
    if max_length <= len(token):
        max_length = len(token)
```

```python
padding_repr = pad_sequences(one_hot_repr, maxlen=max_length, padding='pre')
```

```python
padding_repr
```

```python
import torch
```

```python
## Sending the padding repr to embedding layer
english_padding = torch.LongTensor(padding_repr)
```

```python
from torch.nn import Embedding
```

```python
embedding_layer = Embedding(num_embeddings=500, embedding_dim=100)

with torch.no_grad():
    embedding_repr = embedding_layer(english_padding)
```

```python
embedding_repr
```

```python
embedding_repr = embedding_repr.mean(dim=1)
```
